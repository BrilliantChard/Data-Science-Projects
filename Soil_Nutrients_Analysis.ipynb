```python
maximum = df.groupby(['Name', 'Fertility', 'Photoperiod', 'Season', 'Soil_Type']).agg({'Temperature': 'max', 'Rainfall': 'max', "pH": 'max', 'Yield':"max"})
maximum
```

```python
def iqr(columns):
    mean = columns.mean()
    median = columns.median()
    max_val = columns.max()
    min_val = columns.min()
    iqr = columns.quantile(0.75) - columns.quantile(0.25)

    return pd.Series({
        'Mean': mean,
        'Median': median,
        'Max': max_val,
        'Min': min_val,
        'IQR': iqr
    })
```

```python
descriptive_statistics = df.groupby('Soil_Type')['Yield'].apply(iqr).reset_index()
descriptive_statistics
```

```python
descriptive_statistics = df.groupby('Soil_Type')['Yield'].apply(iqr).unstack().reset_index()
descriptive_statistics
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=descriptive_statistics)
```

```python
sns.barplot(data=descriptive_statistics, x='Soil_Type', y='Mean')
```

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 8))
axes = axes.flatten()

stats_columns = ['Mean', 'Median', 'Max', 'Min', 'IQR']

for i, col in enumerate(stats_columns):
    sns.barplot(x='Soil_Type', y=col, data=descriptive_statistics, ax=axes[i])
    axes[i].set_title(f'{col} of Yield by Soil Type')
    axes[i].set_ylabel(col)
    axes[i].set_xlabel('Soil Type')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
